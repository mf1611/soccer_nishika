```python
import json
import pandas as pd
```

```python
year = 2017
div = 'J1'
team_id = 1
```

```python
with open(f"../input/event/event_play_{year}_{div}_{team_id}.json") as f:
    dic = json.load(f)

full_df = pd.DataFrame.from_dict(dic, orient='index')
print(full_df.shape)
full_df.head()
```

```python
ch = '○' #, '▲', '▽', '◆', 'B', '×'
```

```python
full_df.apply(lambda d: d.str.contains(ch))
```

```python
# 各行（各選手）ごとに，各状態の合計
full_df.apply(lambda d: d.str.contains(ch)).sum(axis=1).values
```

```python
tmp = ['rat_full_play', 'rat_out_play', 'rat_in_play', 'rat_inout_play', 'rat_bench_play', 'rat_susp_play']
colnames = ['rat_full_play', 'rat_out_play', 'rat_in_play', 'rat_inout_play', 'rat_bench_play', 'rat_susp_play']
colnames += [i + '_first' for i in tmp]
colnames += [i + '_second' for i in tmp]
characters = ['○', '▲', '▽', '◆', 'B', '×'] * 3
```
