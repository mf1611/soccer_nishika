```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
import pandas_profiling
import os
import pickle

pd.set_option("display.max_columns", None)
# pd.set_option("display.max_rows", 200)
```

```python
DATA_DIR = '../input/'
```

```python
df_2018 = pd.read_csv(DATA_DIR+'train_2018.csv')
print(df_2018.shape)
df_2018.head()
```

```python
df_2017 = pd.read_csv(DATA_DIR+'train_2017.csv')
print(df_2017.shape)
df_2017.head()
```

```python
df_2016 = pd.read_csv(DATA_DIR+'train_2016.csv')
print(df_2016.shape)
df_2016.head()
```

```python
label = 'time_played'

cols_num = [
     'height',
     'weight',
     'nth_year',
     'is_youth',
     'j1_total_num_played',
     'j1_total_scores',
     'j2_total_num_played',
     'j2_total_scores',
     'j3_total_num_played',
     'j3_total_scores',
     'na_total_num_played',
     'na_total_scores',
     'prev3_num_played',
     'prev2_num_played',
     'prev1_num_played',
     'prev3_scores',
     'prev2_scores',
     'prev1_scores',
     'prev3_time_played',
     'prev2_time_played',
     'prev1_time_played'
]

cols_cat = [
    'team',
     'position',
     'birthdate',
     'salary',
     'nationality',
     'prev3_team',
     'prev2_team',
     'prev1_team',
     'prev3_div',
     'prev2_div',
     'prev1_div',
]

features = cols_num + cols_cat + [label]
```

```python
cols_rename = {col: col+"_prev_year" for col in features}
df_train = pd.merge(df_2017[["id"]+features], df_2016[["id"]+features].rename(columns=cols_rename), on="id", how="left")
df_test = pd.merge(df_2018[["id"]+features], df_2017[["id"]+features].rename(columns=cols_rename), on="id", how="outer")
```
